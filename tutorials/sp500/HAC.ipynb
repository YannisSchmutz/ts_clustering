# Hierarchical Agglomerative Clustering (HAC)

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.cluster.hierarchy import dendrogram, linkage

# Apply the default theme
sns.set()

%matplotlib inline
```

## Prepare and display example data

```python
X = np.array([[0.40, 0.53],
             [0.22, 0.38],
             [0.35, 0.32],
             [0.26, 0.19],
             [0.08,  0.41],
             [0.45, 0.30]])
```

```python
labels = range(1, 7)
plt.figure(figsize=(10, 7))
plt.subplots_adjust(bottom=0.1)
plt.scatter(X[:,0],X[:,1], label='True Position')

for label, x, y in zip(labels, X[:, 0], X[:, 1]):
    plt.annotate(
        label,
        xy=(x, y), xytext=(-3, 3),
        textcoords='offset points', ha='right', va='bottom')
plt.show()
```

## Demonstrate HAC using single linkage

```python
linked = linkage(X, 'single')

labelList = range(1, 7)

plt.figure(figsize=(10, 7))
dendrogram(linked,
            orientation='top',
            labels=labelList,
            distance_sort='descending',
            show_leaf_counts=True)
plt.show()
```

## Demonstrate HAC using complete linkage

```python
linked = linkage(X, 'complete')

labelList = range(1, 7)

plt.figure(figsize=(10, 7))
dendrogram(linked,
            orientation='top',
            labels=labelList,
            distance_sort='descending',
            show_leaf_counts=True)
plt.show()
```

## Demonstrate HAC using average linkage

```python
linked = linkage(X, 'average')

labelList = range(1, 7)

plt.figure(figsize=(10, 7))
dendrogram(linked,
            orientation='top',
            labels=labelList,
            distance_sort='descending',
            show_leaf_counts=True)
plt.show()
```
